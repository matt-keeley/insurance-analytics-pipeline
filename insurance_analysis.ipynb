```python
# Import dependencies and load in original dataset
import pandas as pd

data = pd.read_csv('data/medical_insurance.csv')
```

```python
# Introduce synthetic inconsistencies to the data to simulate real-world messy data

# Duplicates 2% of data, randomly sampled
duplicates = data.sample(frac=0.02, random_state=31) 
data_messy = pd.concat([data, duplicates]).reset_index(drop=True)

# Introduce mixed-data types to annual_premium column
symbol_idx = data_messy[data_messy['annual_premium'].notna()].sample(frac=0.02, random_state=17).index
data_messy.loc[symbol_idx, 'annual_premium'] = data_messy.loc[symbol_idx, 'annual_premium'].apply(
    lambda x: f'${x:.2f}'
)

# Introduce white-space to regions
pad_idx = data_messy.sample(frac=0.05, random_state=99).index
data_messy.loc[pad_idx, 'region'] = data_messy.loc[pad_idx, 'region'].apply(
    lambda x: f'  {x}  '
)

# Introduce incorrect ages to simulate errors in data
age_idx = data_messy.sample(frac=0.005, random_state=66).index
data_messy.loc[age_idx, 'age'] = -1
```

```python
# Save messy data to be used in further analysis

data_messy.to_csv('data/messy_data.csv', index=False)
```
